```python
from rake_nltk import Rake
```

```python
from pprint import pprint

r = Rake()

from Api.NewsDataService import NewsDataService
t = NewsDataService().FetchNewsData(10)

for item in t:
    n = item.get_newsContent()

    r.extract_keywords_from_text(n)
    result = r.get_ranked_phrases_with_scores()
    pprint(result)
    print("\n\n\n\n\n\n\n")

```
